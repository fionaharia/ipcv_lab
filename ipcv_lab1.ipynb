```python
import matplotlib.pyplot as plt
import cv2
```

```python
IMG_PATH="nature.jpg"
img=cv2.imread(IMG_PATH,1)
```

```python
plt.imshow(img)
```

## Convert the image to grayscale

```python
grey_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img, cmap="gray")
```

## Implementing image cropping functionality

```python
x,y,w,h=100,100,400,400
```

```python
crop_img=img[y:y+h,x:x+w]
```

```python
img_a=cv2.imread("nature.jpg",1)
```

```python
plt.imshow(img_a)
```

## Arithmetic Addition operations on the image

```python
x,y,w,h=100,100,400,400
crop_img_a=img_a[y:y+h,x:x+w]
```

## Arithmetic Subtraction operations on the image

```python
add_img=cv2.add(crop_img,crop_img_a)
plt.imshow(add_img)
```

```python
sub_img=cv2.subtract(crop_img,crop_img_a)
plt.imshow(sub_img)
```

## Aritmetic multiplication operations

```python
mul_img=cv2.multiply(crop_img,crop_img_a)
plt.imshow(mul_img)
```

## Aritmetic division operations

```python
div_img = cv2.divide(crop_img, crop_img_a)
plt.imshow(div_img)
```

## Perform arithmetic blending operations on the images

```python
alpha = 0.5
beta = 0.5
blend_img = cv2.addWeighted(crop_img, alpha, crop_img_a, beta, 0)
plt.imshow(blend_img)
```

## Implementing AND, OR, NOT, XOR logical operations on the images

```python
bitwise_and=cv2.bitwise_and(crop_img,crop_img_a, mask=None)
```

```python
bitwise_or=cv2.bitwise_or(crop_img,crop_img_a, mask=None)
```

```python
bitwise_not=cv2.bitwise_not(crop_img, mask=None)
```

```python
bitwise_xor=cv2.bitwise_xor(crop_img,crop_img_a, mask=None)
```

```python
fig=plt.figure(figsize=(20,5))

plt.subplot(1,3,1)
plt.imshow(crop_img)
plt.title('Image1')

plt.subplot(1,3,2)
plt.imshow(crop_img_a)
plt.title('Image2')

fig.add_subplot(1,3,3)
plt.imshow(bitwise_and)
plt.axis('off')
plt.title("A and B")
```

```python
num_bits = 2
left_shifted_img = cv2.bitwise_and(crop_img, 255 << num_bits)

plt.imshow(left_shifted_img)
plt.title("Left Shifted Image")
plt.show()
```

```python
num_bits = 2
right_shifted_img = crop_img >> num_bits

plt.imshow(right_shifted_img)
plt.title("Right Shifted Image")
plt.show()
```
