# **Importing Libraries**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# **Reading a Low Constrast Black and White Image**

```python
IMG_PATH = "elephant.jpg"
img = cv2.imread(IMG_PATH,1)
plt.imshow(img)
```

```python
def show_images(before, after, title):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.title(f'Before {title}')
    plt.imshow(before, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title(f'After {title}')
    plt.imshow(after, cmap='gray')
    plt.axis('off')
    plt.show()
```

# **Apply Constrast Stretching**

```python
def constrast_str(img):
    min = np.min(img)
    max = np.max(img)
    new_img = ((img - min)*(255) / (max - min))
    return new_img
```

```python
constrast_img = constrast_str(img)
show_images(img, constrast_img, 'Constrast Stretching')
```

# **Apply Log Transformation**

```python
def log_transformation(img):
    c = 255 / np.log(1 + np.max(img))
    log_img = c * np.log(1 + img)
    return np.array(log_img, dtype=np.uint8)
```

```python
log_img = log_transformation(img)
show_images(img, log_img, 'Log Transformation')
```

# **Apply Power Law Transformation**

```python
def power_law(img):
  gamma = 2.0
  power_img = ((img / 255) ** gamma * 255).astype(np.uint8)
  return power_img
```

```python
power_law_img = power_law(img)
show_images(img, power_law_img, 'Power Law Transformation')
```

# **Reading a low constrast RGB image**

```python
img = cv2.cvtColor(cv2.imread('full-after.jpg'), cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

# **Applying Contrast Stretching, Log Transformation and Power Law Transformation on the above colored image**

```python
constrast_img = constrast_str(img)
show_images(img, constrast_img, 'Constrast Stretching')

log_img = log_transformation(img)
show_images(img, log_img, 'Log Transformation')

power_law_img = power_law(img)
show_images(img, power_law_img, 'Power Law Transformation')
```

# **Observation**

In this experiment, we applied several image enhancement techniques to a low contrast black and white image and colored image. We observed that:

1. **Contrast Stretching** significantly improved the visibility of the image, particularly by expanding the range of pixel intensity values, making the details more noticeable.

2. **Log Transformation** enhanced the darker regions of the image, bringing out details that were initially hard to see. This transformation was particularly useful in revealing subtle details in the low-intensity areas.

3. Power Law Transformation fine-tunes the brightness and contrast in a more controlled manner. By adjusting the gamma value, we were able to either brighten or darken the image, which provided flexibility
